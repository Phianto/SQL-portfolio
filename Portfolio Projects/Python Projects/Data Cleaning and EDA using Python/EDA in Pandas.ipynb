```python
df.corr(numeric_only=True)  # To find the correlation between the columns

# We can see that the correlation between population and area,density etc. are not very much correlated 
# as the data difference between the columns are much greater in terms of some countries as compared to population column
```

```python
sns.heatmap(df.corr(numeric_only=True), annot = True)  # To get the heatmap of the correlation between each column

plt.rcParams['figure.figsize'] = (20,7)  # To alter the size of the heatmap
plt.show()    
```

```python
df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population",ascending = False)
# Sorting the mean of dataset with 2022 population table descending

```

```python
df.columns  # To gather  all the columns present in the dataset
```

```python
df[df['Continent'].str.contains('Oceania')]  # To find what countries are associated with the continent Oceania
```

```python
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean(numeric_only=True).sort_values(by="2022 Population",ascending = False)
df2

# To select only population data with respect to continent for plotting

## We can also use indexing ('Continent')[df.columns[5:3]].mean()
```

```python
df2.plot()  # we get the line chart but the population seems indistinguishable so we'll transpose the table
```

```python
df3 = df2.transpose()
df3
# transposing the table and saving on a seperate data frame
```

```python
df3.plot()  # getting the line plot which easily demonstrates the exponential population growth in Asia
```

```python
df.boxplot(figsize = (20,10))  
# Using outliners we can see if there is any anomaly in this case any country having solely huge population
# In the below plot as the 2 dots represent china and india in 2022 and on
```
